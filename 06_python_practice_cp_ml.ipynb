```python
""" 
map → transform
filter → select
reduce → aggregate
"""


# *****
# 1. Square each element in a list using map.  
arr = [1,2,3,4,5,6,7,8,9]
print(list(map(lambda x:x**2,arr)))


# 2. Filter even numbers using filter.  
print(list(filter(lambda x:x%2==0,arr)))



""" 
map → 1 input → 1 output (each element)
filter → 1 input → True/False
reduce → 2 inputs → 1 final output
"""
# *****
# 3. Sum a list using reduce.  
from functools import reduce
print(reduce(lambda x,y: x+y,arr))


# ****
# 4. Find max in a list using reduce.
print(reduce(lambda a,b: a if a>b else b,arr))


# 5. Convert strings to ints using map.  
arr = ['1','2','3','4','5','6','7','8']
print(arr)
print(list(map(int,arr)))


# *****
# 6. Filter words longer than 3 letters. 
arr = "My name is Yasin Arafat. What are you doing?".split(" ")
print(list(filter(lambda x: len(x)>3 ,arr)))
 
 
# *****
# 7. Multiply all elements using reduce.  
arr = [1,2,3,4,5,6,7,8,9]
print(reduce(lambda x,y: x*y,arr))


# *****
# 8. Sort a list by lambda key.  
arr = [1,2,3,4,9,10,-5,-6,-9]
print(sorted(arr,key=lambda x:x))


# *****
# 9. Filter primes using filter.  
print(list(filter(lambda x:x>1 and all(x%i!=0 for i in range(2,int(x**0.5+1))),arr)))


# 10. Map lambda to add 1 to each.  
print(list(map(lambda x:x+1,arr)))



```
