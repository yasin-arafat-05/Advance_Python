````python
""" 
map ‚Üí transform
filter ‚Üí select
reduce ‚Üí aggregate
"""


# *****
# 1. Square each element in a list using map.  
arr = [1,2,3,4,5,6,7,8,9]
print(list(map(lambda x:x**2,arr)))


# 2. Filter even numbers using filter.  
print(list(filter(lambda x:x%2==0,arr)))



""" 
map ‚Üí 1 input ‚Üí 1 output (each element)
filter ‚Üí 1 input ‚Üí True/False
reduce ‚Üí 2 inputs ‚Üí 1 final output
"""
# *****
# 3. Sum a list using reduce.  
from functools import reduce
print(reduce(lambda x,y: x+y,arr))


# ****
# 4. Find max in a list using reduce.
print(reduce(lambda a,b: a if a>b else b,arr))


# 5. Convert strings to ints using map.  
arr = ['1','2','3','4','5','6','7','8']
print(arr)
print(list(map(int,arr)))


# *****
# 6. Filter words longer than 3 letters. 
arr = "My name is Yasin Arafat. What are you doing?".split(" ")
print(list(filter(lambda x: len(x)>3 ,arr)))
 
 
# *****
# 7. Multiply all elements using reduce.  
arr = [1,2,3,4,5,6,7,8,9]
print(reduce(lambda x,y: x*y,arr))


# *****
# 8. Sort a list by lambda key.  
arr = [1,2,3,4,9,10,-5,-6,-9]
print(sorted(arr,key=lambda x:x))


# *****
# 9. Filter primes using filter.  
print(list(filter(lambda x:x>1 and all(x%i!=0 for i in range(2,int(x**0.5+1))),arr)))


# 10. Map lambda to add 1 to each.  
print(list(map(lambda x:x+1,arr)))



````

````python

# Implement a queue using deque.   (Queue--> FIFO)

from collections import deque
class Deque:
    def __init__(self):
        self.items = deque()
        
    # insert element to the front:
    def enqueue(self,val):
        self.items.append(val)
        
    # remove from back:
    def dequeue(self):
        if not self.items:
            raise IndexError("Can't remove value from empty deque.")
        return self.items.popleft()
    
    
    # pick the front:
    def front(self):
        if not self.items:
            raise IndexError("Can't show pick from a empty deque")
        return self.items[0]
    
    # is empty:
    def is_empty(self):
        return len(self.items) == 0
    
    # len:
    def size(self):
        return len(self.items)
    
    
d = Deque()
d.enqueue(1)
d.enqueue(2)
d.enqueue(3)
    
print("is_empty:",d.is_empty())
print("size:",d.size())
print("front:",d.front())
print("remove:",d.dequeue())
print("front:",d.front())

````

### **Problem 1: Balanced Parentheses**

You are given a string `s` consisting only of the characters
`(`, `)`, `{`, `}`, `[`, `]`.

Your task is to determine whether the string is **balanced**.

A string is considered balanced if:

* Every opening bracket has a corresponding closing bracket
* Brackets are closed in the correct order

#### Input

A single string `s`

#### Output

Print `"YES"` if the string is balanced, otherwise print `"NO"`

#### Example

```
Input:
({[]})

Output:
YES
```

---


````python


from collections import deque
input_str = "({[]})"
stack = deque()
mp = {
    ')':'(',
    '}':'{',
    ']':'['
}

for ch in input_str: 
    
    # opening bracket push to the stack:
    if ch in ['(','{','[']:
        stack.append(ch)
        
    # if closing then pop from stack mp for safety:
    else:
        if not mp[ch]: 
            break 
        stack.pop()
        
if stack:
    print("NO")
else:
    print("YES")
        
        
        
````

### **Problem 4: Simple Queue Simulation**

You are given `Q` operations to perform on an initially empty queue.

Each operation is one of the following:

* `push x` ‚Üí insert integer `x` into the queue
* `pop` ‚Üí remove the front element of the queue (ignore if queue is empty)
* `front` ‚Üí print the front element (print `-1` if empty)
* `size` ‚Üí print the number of elements in the queue

#### Input

First line contains integer `Q`
Next `Q` lines contain operations

#### Output

Print output for each `front` or `size` operation

````python



from collections import deque

class Queue:
    def __init__(self):
        self.items = deque()
    
    def push(self, x):
        self.items.append(x)
    
    def pop(self):
        if self.items:
            return self.items.popleft()
        return None
    
    def front(self):
        if self.items:
            return self.items[0]
        return None
    
    def size(self):
        return len(self.items)

q = Queue()
q.push('a')
q.push('b')
q.push('c')

print(q.front())  # a
print(q.pop())    # a
print(q.front())  # b
print(q.size())   # 2

````

### **Problem 5: Round-Robin Scheduling**

You are given `N` processes arranged in a queue.
Each process has:

* a process ID
* a burst time

The CPU processes each task for **exactly `T` units of time** per turn (time quantum).

If a process‚Äôs remaining burst time is greater than `T`, it is placed back at the end of the queue with reduced burst time.

Otherwise, the process completes and is removed from the queue.

#### Input

* Integer `N` (number of processes)
* Integer `T` (time quantum)
* Next `N` lines: `process_id burst_time`

#### Output

Print the **order in which processes complete**

#### Example

```
Input:
3 4
P1 8
P2 5
P3 2

Output:
P3 P1 P2
```


````python
""" 
** Queue ‡¶¶‡¶ø‡ßü‡ßá ‡¶è‡¶ï‡¶ü‡¶æ‡¶∞ ‡¶™‡¶∞ ‡¶è‡¶ï‡¶ü‡¶æ process ‡¶ö‡¶æ‡¶≤‡¶æ‡¶¨‡ßá
** ‡¶Ø‡ßá‡¶ü‡¶æ‡¶∞ burst_time ‡¶∂‡ßá‡¶∑ ‡¶π‡¶¨‡ßá ‚Üí output ‡¶è ‡¶Ø‡¶æ‡¶¨‡ßá
‡¶®‡¶æ ‡¶∂‡ßá‡¶∑ ‡¶π‡¶≤‡ßá ‚Üí ‡¶Ü‡¶¨‡¶æ‡¶∞ queue-‡¶∞ ‡¶∂‡ßá‡¶∑‡ßá ‡¶Ø‡¶æ‡¶¨‡ßá
"""

from collections import deque
input_str = "3 4 P1 8 P2 5 P3 2".split(" ")
idx = 0 
N = int(input_str[idx]);idx+=1
T = int(input_str[idx]);idx+=1
q = deque()
for _ in range(N):
   process = input_str[idx];idx+=1 
   burst_time = int(input_str[idx]);idx+=1
   q.append([process,burst_time])
   
   
completed = []
while q:
   pid, burst = q.popleft()
   if burst<=T:
      completed.append(pid)
   else:
      burst -=T 
      q.append([pid,burst])
      
print(" ".join(completed))
    
````

## **Problem 1: Next Greater Element (Right Side)**

If no such element exists, print `-1`.

### Input

* Integer `n`
* Array of `n` integers

### Output

* Print `n` integers representing the next greater element for each position

### Example

```
Input:
5
4 5 2 25 7

Output:
5 25 25 -1 -1
```

````python

from collections import deque 

n = 5
arr = [4,5,2,25,7]
q = deque()

for val in arr:
    q.append(val)


ans = []
idx = 0 

while q:
    ele = q.popleft()
    for val in q:
        if val>ele:
            ans.insert(idx,val);idx+=1
            break
        else:
            ans.insert(idx,-1)
            
print(ans)

# Complexity: O(n^2), TLE CP 
````
